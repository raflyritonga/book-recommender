```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
import pandas as pd
```

```python
books = pd.read_csv("../dataset/books_cleaned.csv")
books
```

```python
books['tagged_description'].to_csv("../dataset/tagged_description.txt", 
                            sep='\n',
                            index=False,
                            header=False)
```

```python
raw_documents = TextLoader("../dataset/tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en", model_kwargs={"device": "cuda"})
db_books = Chroma.from_documents(
    documents,
    embedding=embeddings
)
```

```python
query = "a book about a young who discovers they have magical powers"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books['isbn13'] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommandations(
    query: str, 
    top_k: int = 10
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)
    books_list = []
    for i in range (0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books['isbn13'].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommandations("a book about to teach chilcren about nature")
```
